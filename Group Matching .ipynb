```python
import numpy as np
import pandas as pd
import random

from faker import Faker
fake = Faker()
from faker.providers.person.en import Provider
```

```python
#creating function for generating dataset

def random_names(name_type, size):
    """
    Generate n-length ndarray of person names.
    name_type: a string, either first_names or last_names
    """
    names = getattr(Provider, name_type)
    return np.random.choice(names, size=size)

def random_genders(size, p=None):
    """Generate n-length ndarray of genders."""
    if not p:
        # default probabilities
        p = (0.5, 0.5)
    gender = ("M", "F")
    return np.random.choice(gender, size=size, p=p)

def random_program(size, p=None):
    """Generate n-length ndarray of program."""
    if not p:
        # default probabilities
        p = (0.5, 0.5)
    program = ("Masters", "Bachelors")
    return np.random.choice(program, size=size, p=p)
```

```python
size = 50  
df = pd.DataFrame(columns=['FirstName', 'LastName', 'Gender'])
df['FirstName'] = random_names('first_names', size)
df['LastName'] = random_names('last_names', size) 
df['Gender'] = random_genders(size)
df['Program'] = random_program(50)
```

```python
Age = []
for i in range(0,50):
    x = random.randint(18,40)
    Age.append(x)
    
se = pd.Series(Age)
df['Age'] = se.values
```

```python
#list of interests and skills to be used 

I = ['Sports','Art','VolunteerWork','Travel','Blogging','Gaming','Music']

S = ['Communication','Creativity','ProblemSolving','Teamwork','Mathematics','Management','Troubleshooting','Technical','Research']
```

```python
ie = []
for i in range(0,50):
    x = random.sample(I, k=3)
    ie.append(x)
    
se = pd.Series(ie)
df['Interests'] = se.values

ir = []
for i in range(0,50):
    x = random.sample(S, k=3)
    ir.append(x)
    
ss = pd.Series(ir)
df['Skills'] = ss.values
```

# Using fuzzywuzzy for string comparision, and finding group of rows having similar interests

```python
from fuzzywuzzy import fuzz

from fuzzywuzzy import process
name = 'Travel Art'

def func(name):
    matches = df.apply(lambda row: (fuzz.token_sort_ratio(row['Interests'], name) >= 88), axis=1)
    return [i for i, x in enumerate(matches) if x]
```
